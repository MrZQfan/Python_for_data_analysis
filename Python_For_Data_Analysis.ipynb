        需要注意的是：数组切片是原始数据的视图，这意味着数据不会被复制，视图上的任何修改都会直接反应到数组上,如果你想得到的是ndarray切片的一份副本而不是视图，就需要显式地进行复制操作，例如arr_1[2:4].copy() 
        由于NumPy的设计目的是处理大数据，所以可以想象，如果NumPy坚持要将数据复制来复制去的话会产生何等的性能和内存问题。

## 逻辑运算

### 将条件逻辑表述为数组运算

    numpy.where函数是三元表达式 x if condition else y 的矢量化版本

```python
xarr = np.array([1.1, 1.2, 1.3, 1.4, 1.5])
yarr = np.array([2.1, 2.2, 2.3, 2.4, 2.5])
cond = np.array([True, False, True, True, False])
reslut = np.where(cond, xarr, yarr)
reslut
```

np.where的第二个和第三个参数不必是数组，他们都可以是标量。在数据分析工作中，where通常用于根据另一个数组而产生一个新的数组

```python
#有一个随机数据组成的矩阵，希望将所有正值替换为2，将所有负值替换为-2
arr = np.random.randn(4,4)
np.where(arr > 0, 2, -2)
```

```python
np.where(arr > 0, 2, arr)
```

## 用于数组的文件输入与输出

### 将数组以二进制格式保存到磁盘

np.save和np.load是读写磁盘数组数据的两个主要函数。默认情况下，数组是以未压缩的原始二进制格式保存在扩展名为.npy的文件中的。

```python
#保存
arr = np.arange(10)
np.save('some_array',arr) #不用加.npy，保存文件时会自动添加
```

对于任何单字符分隔符文件，可以直接使用Python内置csv模块。将任意已打开的文件或文件型对象传给csv.reader：

```python
import csv
f = open('ex6.csv')
reader = csv.reader(f)
for line in reader:
    print (line)
    
```

现在，为了使数据格式合乎要求，来对其做一些整理工作：

```python
lines = list(csv.reader(open('ex6.csv')))
header, values = lines[0], lines[1:]
data_dict = {h: v for h, v in zip(header, zip(*values))}
data_dict
```

CSV文件的形式有很多。只需定义csv.Dialect的一个子类即可定义出新格式（如专门的分隔符、字符串引用约定、行结束符等）：

```python
class my_dialect(csv.Dialect):
    lineterminator = '\n'
    delimiter = ';'
    quotechar = '"'
    quoting = 0
    
reader = csv.reader(f, dialect = my_dialect)
reader
```

```python
reader = csv.reader(f, delimiter = '|')
reader
```

    CSV的语支选项：
    delimiter：用于分隔字段的单字符字符串，默认为“，”
    lineterminator：用于写操作的行结束符，默认为“\r\n”。读操作将忽略此选项，它能认出跨平台的行结束符
    quotechar：用于带有特殊字符（如分隔符）的字符引用符号。默认为“"”
    quoting：引用约定。详细看文档
    skipinitialspace：忽略分隔符后面的空白符。默认为False
    doublequote：如何处理字段内的引用符号，如果为True，则双写。详细看文档
    escapechar：用于对分隔符进行转义的字符串（如果quoting被设置为csv.QUOTE_NONE的话）。默认禁用

注意：对于那些使用复杂分隔符或多字符的分隔符的文件，csv就无能为力了。这种情况下，你就只能使用字符串的split方法或正则表达式方法re.split进行拆分和其他整理工作了。

要手工输入分隔符文件，你可以使用csv.writer。它接受一个已打开且可写的文件对象以及跟csv.reader相同的那些语支和格式化选项：

说明：parse()方法用来从文件或者类文件对象中解析，返回的是一个ElementTree对象，而不是字符串解析方法的Element对象。ElementTree类：一个ElementTree主要是围绕在一个有根节点的树的文档包装类。它提供了很多方法来解析，序列化以及一般的文档处理。一个最大的区别是它作为一个整体文档来序列化。与之相对的是序列化成单个的元素。用parse函数从文件中解析xml，解析得到的是一个ElementTree的实例，用这个实例的 getroot函数就能得到xml中的root。root是对象Element的一个实例，对这个root可以做indexing，即用过root[n] 可以得到root下相应的子节点，这些子节点同样也是Element的实例，所以通过root[n][m]就可遍历各个节点。

现在，从文档中找到正确表格的办法就是反复试验了。

有些网站会给目标表格加上一个id属性：

```python
tables = doc.findall('.//table')
tables
```

```python
table_1 = tables[0]
table_2 = tables[1]
rows = table_2.findall('.//tr') #每个表格都有一个标题行，然后才是数据行
rows
```

对于标题行和数据行，我们希望获取每个单元格内的文本。对于标题行，就是th单元格，对于数据行就是td单元格。< th >和< td >标签都是用于表格单元格显示的。不同的是< th >定义表格内的表头单元格，且单元格中加粗显示。

```python
def _unpack(row, kind = 'td'):
    elts = row.findall('.//%s' % kind)
    return [val.text_content() for val in elts]
```

```python
_unpack(rows[0], kind = 'th') #标题行
```

```python
_unpack(rows[1], kind = 'td') #数据行1
```

如此以来，我们就得到了想要的数据，虽然上面只是1行

现在，把所有的步骤综合起来，将数据转化成一个DataFrame。且由于数值型数据仍然是字符串格式，所以我们希望将部分列（可能不是全部）转换问浮点数格式，其中用到了pandas中的一个TextParser类可以自动类型转换（read_csv和其他解析函数其实在内部都用到了它）:

```python
from pandas.io.parsers import TextParser
```

```python
def parse_options_data(table):
    rows = table.findall('.//tr')
    header = _unpack(rows[0], kind = 'th')
    data = [_unpack(r) for r in rows[1:]]
    return TextParser(data, names = header).get_chunk() #get_chunk方法,它使你可以读取任意大小的块
```

### 数据库风格的DataFrame合并

数据集的合并（merge）或连接（join）运算是通过一个或多个键将行链接起来的。这些运算是关系型数据库的核心。pandas的merge函数是对数据应用这些算法的主要切入点。

```python
import pandas as pd
from pandas import DataFrame,Series
df1 = DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'], 'data1': range(7)})
df2 = DataFrame({'key': ['a', 'b', 'd'], 'data2': range(3)})
df1
```

```python
df2
```

df1和df2的合并：（默认内联接）

```python
pd.merge(df1, df2)
```

说明：没有指定哪个列进行连接的话，merge就会将重叠列的列名当做键。不过最好显示指定一下。

指定列名进行连接：

```python
pd.merge(df1, df2, on = 'key')
```

如果想要连接的两个对象的列名不同，也可以分别进行指定：

```python
df3 = DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'], 'data1': range(7)})
df4 = DataFrame({'rkey': ['a', 'b', 'd'], 'data2': range(3)})
pd.merge(df3, df4, left_on = 'lkey', right_on = 'rkey')
```

```python
pd.merge(df4, df3, left_on = 'rkey', right_on = 'lkey')
```

说明：左表列和右表列要与左表和右表相对应，顺序也不能错，不然会报错，例pd.merge(df3, df4, left_on = 'rkey', right_on = 'lkey')就不对

不难发现pd.merge的联接方式相当于数据库中的内联接：

```python
pd.merge(df1, df2, how = 'inner') 
```

说明：how默认就是内联接，内联接求取的是连接键的交集  
当然也有左联接、右联接和外联接，外联接求取的是键的并集，组合了左联接和右联接的效果：

```python
pd.merge(df1, df2, how = 'outer')
```

```python
pd.merge(df1, df2, how = 'left')
```

    left：参与合并的左侧DataFrame
    right：参与合并的右侧DataFrame
    how：“inner”、“outer”、“left”、“right”之一。默认为“inner”
    on：用于连接的列名。必须存在于左右两个DataFrame对象中。如果未指定，且其他连接键也未指定，则以left和right列名的交集作为连接键
    left_on：左侧DataFrame中用作连接键的列
    right_on：右侧DataFrame中用作连接键的列

    suffixes：字符串值元组，用于追加到重叠列名的末尾，默认为('_x', '_y')。例如，如果左右两个DataFrame对象都有“data”，则结果中就会出现“data_x”和“data_y”

### 离散化和面元划分

为了便于分析，连续数据常常被离散化或拆分成“面元”（bin）。

#### cut

假设有一组人员数据，希望将它们划分为不同的年龄组：

```python
ages = [20, 22, 25, 27, 21, 23, 37, 31, 61, 45, 41, 32]
```

将数据划分为：“18到25”、“26到35”、“35到60”以及“60以上”几个面元：  
要实现该功能，你需要使用pandas的cut函数：

```python
bins = [18, 25, 35, 60, 100]
cats = pd.cut(ages, bins)
cats
```

说明：pandas返回的是一个特殊的Categorical对象，可以将其看做一组表示面元名称的字符串。

```python
cats.codes #ages所在分组
```

```python
pd.value_counts(cats)
```

与数学中的区间一样，圆括号表示开，方括号表示闭，而决定哪边是闭端可以通过right = False来修改：

```python
pd.cut(ages, [18, 26, 36, 61, 100], right = False)
```

也可以设置自己的面元名称，将labels选项设置为一个列表或数组即可：

```python
group_names = ['Youth', 'YoungAdult', 'MiddleAged', 'Senior']
pd.cut(ages, bins, labels = group_names)
```

```python
cats = pd.cut(ages, bins, labels = group_names)
pd.value_counts(cats)
```

说明：上面那个像小于关系的东西好像出了点问题，但是结果并未出错

如果向cut传入的是面元的数量而不是确切的面元边界，则它会根据数据的最小值和最大值计算等长面元：

```python
data = np.random.rand(20)
pd.cut(data, 4, precision = 2)
```

说明：precision = 2指的是精确度

#### qcut

qcut是一个非常类似于cut的函数，它可以根据样本分位数对数据进行面元划分。根据数据的分布情况，cut可能无法使各个面元中含有相同数量的数据点。而qcut由于使用的是样本分位数，因此可以得到大小基本相等的面元：

```python
data = np.random.randn(1000)  #正态分布
cats = pd.qcut(data, 4)  #按四分位数进行切割
cats
```

```python
pd.value_counts(cats)
```

与cut相同，可以设置自定义的分位数（0到1之间的数值，包含端点）：

```python
pd.qcut(data, [0, 0.1, 0.5, 0.9, 1])
```

    count：返回子串在字符串中的出现次数（非重叠）

    replace：用另一个字符串替换指定子串
    strip、rstrip、lstrip：去除空白符（包括换行符）。相当于对各个元素执行x.strip()（以及rstrip、lstrip）
    split：通过指定的分隔符将字符串拆分成一组子串
    lower、upper：分别将字母字符转换为小写或大写
    ljust、rjust：用空格（或其他字符）填充字符串的空白侧以返回符合最低宽度的字符串

现在，我们想根据分类对数据做一些分析或图形化工作，但各个分类字段中可能含有多个分类。此外，各个分类信息不仅有一个编码，还有一个英文名称（可能还有一个法语名称）。因此需要对数据做一些规整化处理。

首先，定义函数，分别用于获取所有分类的列表，以及将各个分类信息拆分为编码和英语名称：

```python
def to_cat_list(catstr):
    stripped = (x.strip() for x in catstr.split(','))
    return [x for x in stripped if x]

def get_all_categories(cat_series):
    cat_sets = (set(to_cat_list(x)) for x in cat_series)
    return sorted(set.union(*cat_sets))

def get_english(cat):
    code, names = cat.split('.')
    if '|' in names:
        names = names.split(' | ')[1]
    return code,names.strip()
```

```python
import numpy as np
all_codes = get_code(all_cats)
code_index = pd.Index(np.unique(all_codes))
dummy_frame = DataFrame(np.zeros((len(data),len(code_index))), index = data.index, columns = code_index)
```

```python
def basic_haiti_map(ax = None, lllat = 17.25, urlat = 20.25, lllon = -75, urlon = -71):
    m = Basemap(ax = ax, projection = 'stere',
                lon_0 = (urlon + lllon)/2,
                lat_0 = (urlat + lllat)/2,
                llcrnrlat = lllat, urcrnrlat = urlat,
                llcrnrlon = lllon, urcrnrlon = urlon,
                resolution = 'f') #创建极球面投影的Basement实例
    #绘制海岸线、州界、国界以及地图边界
    m.drawcoastlines()
    m.drawstates()
    m.drawcountries()
    return m
```

# 数据聚合与分组运算

## GroupBy技术

Hadley Wickham创造了一个用于表示分组运算的术语“split-apply-combine”（拆分-应用-合并），很好的描述了整个过程。

分组运算的第一个阶段，pandas对象中的数据会根据你所提供的一个或多个键被拆分（split）为多组。拆分操作是在对象的特定轴上执行的，例如DataFrame可以在其行（axis = 0）或列（axis = 1）上进行分组。然后将一个函数应用（apply）到各个分组并产生一个新值。最后，所有这些函数的执行结果会被合并（combine）到最终的结果对象中。结果对象的形式一般取决于数据上执行的操作。

### groupby方法

分组键可以由多种形式，且类型不必相同  
1.列表或数组，其长度与待分组的轴一样  
2.表示DataFrame某个列名的值  
3.字典或Series，给出待分组轴上的值与分组名之间的对应关系  
4.函数，用于处理轴索引或索引中的各个标签  
需要注意的是，后三种都只是快捷方式而已，其最终目的仍然是产生一组用于拆分的对象。

```python
import numpy as np
from pandas import Series, DataFrame
import pandas as pd
```

```python
df = DataFrame({'key1': ['a', 'a', 'b', 'b', 'a'], 'key2': ['one', 'two', 'one', 'two', 'one'],
                'data1': np.random.randn(5), 'data2': np.random.randn(5)})
df
```

按key1进行分组，并计算data1列的平均值：  
采用方法：访问data1，并根据key1调用groupby：

```python
grouped = df['data1'].groupby(df['key1'])
grouped
```

说明：grouped是一个GroupBy对象，它实际上还没有进行任何计算，只是含有一些有关分组键df['key']的中间数据而已。换句话说，该对象已经有了接下来对各分组执行计算所需的一切信息。

于是，我们调用GroupBy对象的mean方法来计算平均值：

```python
grouped.mean()
```

以上过程说明：  
数据（Series）根据分组键进行了聚合，产生了一个新的Series，其索引为key1列中的唯一值。之所以结果中索引的名称为key1，是因为沿用了DataFrame的列df['key']的名字。

若一次传入多个数组，那么：

```python
means = df['data1'].groupby([df['key1'], df['key2']]).mean()
means
```

说明：这里通过两个键对数据进行了分组，得到的Series具有一个层次化索引（由唯一的键组成的）

```python
means.unstack()
```

分组键除了为上述的Series外，还可以是任何长度适当的数组：

```python
states = np.array(['Ohio', 'California', 'California', 'Ohio', 'Ohio'])
years = np.array([2005, 2005, 2006, 2005, 2006])
df['data1'].groupby([states, years]).mean()
```

此外，还可以将列名（可以是字符串、数字或其他Python对象）用作分组键：

```python
df.groupby('key1').mean()
```

```python
df.groupby(['key1', 'key2']).mean()
```

使用用groupby时，经常会用到GroupBy对象的size方法，它可以返回一个含有分组大小的Series：

```python
df.groupby(['key1', 'key2']).size()
```

### 对分组进行迭代

GroupBy对象支持迭代，可以产生一组二元元组（由分组名和数据块组成）

```python
for name, group in df.groupby('key1'):
    print(name)
    print(group)
```

对于多重键的情况下，元组的第一个元素将会是由键值组成的元组：

```python
for (k1, k2), group in df.groupby(['key1', 'key2']):
    print(k1, k2)
    print(group)
```

说明：语法糖指计算机语言中添加的某种语法，这种语法对语言的功能并没有影响，但是更方便程序员使用。通常来说使用语法糖能够增加程序的可读性，从而减少程序代码出错的机会。

例如，计算data2列的平均值并以DataFrame形式得到结果：

```python
df.groupby(['key1', 'key2'])['data2'].mean()
```

### 通过字典或Series进行分组

除数组外，分组信息还可以其他形式存在。

```python
people = DataFrame(np.random.randn(5, 5), columns = ['a', 'b', 'c', 'd', 'e'], index = ['Joe', 'Steve', 'Wes', 'Jim', 'Travis'])
people.loc[2:3, ['b', 'c']] = np.nan #添加几个NA值
people
```

假设已知列的分组关系，并希望根据分组计算列的总计：

```python
mapping = {'a': 'red', 'b': 'red', 'c': 'blue', 'd': 'blue', 'e': 'red', 'f': 'orange'}
by_column = people.groupby(mapping, axis = 1)
by_column.sum()
```

Series也有同样的功能，它可以被看做一个固定大小的映射。如果用Series作为分组键，则pandas会检测Series以确保其索引跟分组轴是对齐的：

```python
map_series = Series(mapping)
people.groupby(map_series, axis = 1).count()
```

### 通过函数进行分组

相较于字典或Series，Python函数在定义分组映射关系时可以更有创意且更为抽象。任何被当做分组键的函数都会在各个索引值上被调用一次，其返回值就会用作分组名称。

例如，以上一节的示例DataFrame为例，其索引值为人的名字。假设你希望根据人名的长度进行分组，虽然可以求取一个字符串长度的数组，但其实仅仅传入len函数就可以了：

```python
people.groupby(len).sum()
```

将函数跟数组、列表、字典、Series混合使用也不是问题，因为任何东西最终都会被转换为数组：

```python
key_list = ['one', 'one', 'one', 'two', 'two']
people.groupby([len, key_list]).min()
```

## 数据聚合

对于聚合，这里指的是任何能够从数组中产生标量值得数据转换过程。之前的例子中mean、count、min以及sum就在其中。

许多常见的聚合运算都有就地计算数据集统计信息的优化实习，然而，并不只能使用这些方法，你还可以使用自发明的聚合运算，还可以调用分组对象上已经定义 好的任何方法：  
例如，quantitle可以计算Series或DataFrame列的样本分位数：

```python
df
```

```python
grouped = df.groupby('key1')
grouped['data1'].quantile(0.9)
```

```python
df.groupby('key1')['data1'].quantile(0.9)
```

说明：虽然quantile并没有明确的实习与GroupBy，但它是一个Series方法，所以这里是能用的，实际上，GroupBy会高效的队Series进行切片，然后对各片调用piece.quantile（0.9），最后将这些结果组装成最终结果。

自创聚合函数：将其传入aggregate或agg方法即可：

```python
def peak_to_peak(arr):
    return arr.max() - arr.min()

grouped.agg(peak_to_peak)
```

有些严格上来讲并非聚合运算的函数也是可以在这里使用的，例如describe：

```python
grouped.describe()
```

说明：在分组级运算中会说明原理

经过优化的groupby的方法：  
     
     count：分组中非NA值的数量
     sum：非NA值的和
     mean：非NA值的平均值
     median：非NA值的算术中位数
     std、var：无偏（分母为n - 1）标准差和方差
     min、max：非NA值的最小和最大值
     prod：非NA值的积
     first、last：第一个和最后一个非NA值

注意：自定义聚合函数要比经过优化的函数慢得多。这是因为在构造中间分组数据块时存在非常大的开销（函数调用、数据重排等）

### 更高级的聚合功能：面向列的多函数应用

可以看到，对Series或DataFrame列的聚合运算其实就是使用aggregate（使用自定义函数）或调用诸如mean、std之类的方法。那么，若希望对不同的列使用不同的聚合函数，或一次应用读个函数，该如何操作呢？

演示：

1.加载数据

```python
tips = pd.read_csv('PFDA_data/tips.csv')
tips['tip_pct'] = tips['tip']/tips['total_bill'] #添加“小费占总额百分比”
tips[:6]
```

2.根据sex和smoker对tips进行分组：

```python
grouped = tips.groupby(['sex', 'smoker'])
```

3.对于经过优化的groupby的方法，可以将函数名以字符串的形式传入：

```python
grouped_pct = grouped['tip_pct']
grouped_pct.agg('mean')
```

4.一次应用多个函数：若传入一组函数或函数名，得到的DataFrame的列就会以相应的函数命名：

```python
grouped_pct.agg(['mean', 'std', peak_to_peak])
```

4.1 指定列名
你并非一定要接受GroupBy自动给出的列名，特别是lambda函数，它们的名称是'lamda'，这样辨识度就很低了。若传入的是一个由(name, function)元组组成的列表，则个元组的第一个元素就会被用作DataFrame的列名（可以将这种二元元组列表看做一个有序映射）：

```python
grouped_pct.agg([('foo', 'mean'), ('bar', np.std)])
```

5.一个或多个函数同时应用于多个列

```python
functions = ['count', 'mean', 'max']
result = grouped['tip_pct', 'total_bill'].agg(functions)
result
```

说明：结果DataFrame拥有层次化的列，这相当于分别对各列进行聚合，然后用concat将结果组装到一起（列名用作keys参数）

与前相同：同样可以传入带有自定义名称的元组列表：

```python
ftuples = [('Durchschnitt', 'mean'), ('Abweichuang', np.var)]
grouped['tip_pct', 'total_bill'].agg(ftuples)
```

6.不同的列应用不同的函数

方法是向agg传入一个从列名映射到函数的字典：

```python
grouped.agg({'tip': np.max, 'size': 'sum'})
```

扩：其中有列应用的函数是多个：

```python
grouped.agg({'tip_pct': ['min', 'max', 'mean', 'std'], 'size': 'sum'})
```

说明：只有将多个函数应用到至少一列时，DataFrame才会拥有层次化的列

### 以“无索引”的形式返回聚合数据

以上所有示例中的聚合数据都有由唯一的分组键组成的索引（可能还是层次化的）。由于并不总是需要如此，所以可以向groupby传入as_index = False以禁用该功能：

```python
tips.groupby(['sex', 'smoker'], as_index = False).mean()
```

说明：对结果调用reset_index也能得到这种形式的结果；groupby的这种用法比较缺乏灵活性
